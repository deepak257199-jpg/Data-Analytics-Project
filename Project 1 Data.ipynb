```python
import matplotlib.pyplot as plt
x=[1,2,3,3,5]
y=[10,20,35,45,50]

plt.figure(figsize=(4,3)) # figure size
plt.plot(x,y, color='red', marker='o',linestyle='--', linewidth=2,markersize=12)
plt.title("Kuch bhi tittle hai")
plt.xlabel("x-axis label hai")
plt.ylabel("y-axis label hai")
plt.show()
```

```python
import random
data=[random.randint(1,20) for _ in range(500)]
plt.hist(data, bins=20) #histogram chart
plt.title("Histogram example")

plt.show()
```

```python

categories = ['A','B','C','D','E']
sales = [10,20,55,35,20]

plt.pie(sales,labels=categories, autopct='%1.1f%%',startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

```python
y1 = [10,20,25,35,60]
y2 = [20,40,45,55,60]
plt.scatter(y1,y2)
plt.title("Scatter plot example")
plt.show()
```

```python
#data1 bar chart
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [10,20,25,35,60]
#data 2 scatter plot
y1 = [10,20,25,35,60]
y2 = [20,40,45,55,60]

plt.figure(figsize=(10,4))

#first plot bar chart
plt.subplot(1,2,1)
plt.bar(categories,sales)
plt.title("Daily sales")
plt.xlabel("week Days")
plt.ylabel("Sales")

#second plot scatter plot
plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title("User example")
plt.xlabel("User1")
plt.ylabel("User2")
plt.show()
```
